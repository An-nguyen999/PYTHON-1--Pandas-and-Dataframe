# 1.3 Apply()

```python
df.item_price.apply(lambda x: x.replace('$','')).dtype
```

```python
df.item_price = df.item_price.apply(lambda x: float(x.replace('$',''))) #thêm dòng float vào để chuyển thành kiểu dữ liệu float
```

```python
df.head() #ở trên đã gán cột df.item_price nên dữ liệu trong bảng sẽ thay đổi
```

```python
df['total_price'] = df['quantity'] * df['item_price'] #cách thêm 1 cột trong dataframe
```

```python
df.head()
```

```python
revenue = df['total_price'].sum()
revenue
```

# 1.4 Group By

```python
c = df.groupby('item_name').sum('quantity')['quantity']
c.sort_values(ascending = False) #sắp xếp theo thứ tự giám dần, còn nếu tăng dần thì xóa trong ngoặc là ok
```
